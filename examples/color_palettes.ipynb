# Color Palette  

In this section, we will introduce a rich set of color palettes in orbit for visualization including 

* Qualitative Palette for visualizing discrete categorical data 
* Sequential Palette that includes gradient colors of mono hue for numeric and inherently ordered values; it is also colorblind friendly
* Diverging Palette with two hues for inherently ordered values or for numeric variable has a meaningful central value, like zero. The colors gradients from the center to the right and the left.

```python
import seaborn as sns 
from matplotlib import pyplot as plt 
import pandas as pd
import numpy as np

import orbit
from orbit.utils.dataset import load_iclaims
from orbit.eda import eda_plot

import orbit.constants.palette as palette
import inspect

%load_ext autoreload
%autoreload
```

```python
df = load_iclaims()
df['week'] = pd.to_datetime(df['week'])
```

```python
df.head()
```

## how to use a single color from QualitativePalette 

```python
_ = plt.plot(df['week'], df['claims'], color=palette.QualitativePalette.Bar5.value[4])
```

## how to use sequential colors from SequentialPalette


```python
_ = eda_plot.ts_heatmap(df = df, date_col = 'week', value_col='claims', palette=palette.SequentialPalette.Seafoam.value)
```

```python
_ = eda_plot.ts_heatmap(df = df, date_col = 'week', value_col='claims', palette=palette.SequentialPalette.Sunshine.value, normalization=True)
```

## how to use sequential colors from DivergingPalette


```python
var_list = ['trend.unemploy', 'trend.filling', 'trend.job', 'sp500', 'vix']
_ = eda_plot.correlation_heatmap(df, var_list = var_list, fig_width=10, fig_height=8, palette=palette.DivergingPalette.Unclesam.value)
```

```python
var_list = ['trend.unemploy', 'trend.filling', 'trend.job', 'sp500', 'vix']
_ = eda_plot.correlation_heatmap(df, var_list = var_list, fig_width=10, fig_height=8, palette=palette.DivergingPalette.Rainbow.value)
```

## Also compatible with matplotlib named colors, check documentation [here](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
_ = eda_plot.dual_axis_ts_plot(df=df, var1='trend.unemploy', var2='claims', date_col='week', 
                               color1='tab:blue', color2='tab:orange')
```
